```python
import urllib.request
import json
import geopandas as gpd
import matplotlib.pyplot as plt

# API call to retrieve GeoJSON data
try:
    url = "https://gateway.apiportal.ns.nl/Spoorkaart-API/api/v1/spoorkaart"
    
    hdr = {
        'Cache-Control': 'no-cache',
        'Ocp-Apim-Subscription-Key': '2e586ae40a434c3f96c891fb53b8e54e',  # Your subscription key
    }

    # Make the request
    req = urllib.request.Request(url, headers=hdr)
    response = urllib.request.urlopen(req)

    # Check for successful response
    if response.getcode() == 200:
        print(f"API Response Code: {response.getcode()} - Success")

        # Read the raw response data
        response_data = response.read().decode('utf-8')
        
        # Parse the JSON response
        geojson_data = json.loads(response_data)
        
        # Access the 'features' within 'payload'
        if 'payload' in geojson_data and 'features' in geojson_data['payload']:
            features = geojson_data['payload']['features']
            
            # Use geopandas to read the GeoJSON features into a GeoDataFrame
            gdf = gpd.GeoDataFrame.from_features(features)
            
            # Set the CRS for the GeoDataFrame, if not already set
            gdf.set_crs(epsg=4326, inplace=True)  # Assuming EPSG:4326 for the GeoJSON data

            # Load the Netherlands shapefile from Natural Earth dataset
            netherlands_shapefile_path = "ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp"
            world = gpd.read_file(netherlands_shapefile_path)

            # Filter the GeoDataFrame to get only the Netherlands
            netherlands = world[world['NAME'] == "Netherlands"]

            # Ensure both GeoDataFrames are in the same coordinate reference system (CRS)
            if gdf.crs != netherlands.crs:
                gdf = gdf.to_crs(netherlands.crs)

            # Filter the GeoDataFrame to include only features within the Netherlands
            gdf_filtered = gdf[gdf.geometry.intersects(netherlands.unary_union)]

            # Plot the Netherlands outline
            ax = netherlands.plot(figsize=(10, 10), color='lightgrey', edgecolor='black')

            # Plot the filtered GeoDataFrame on top of the Netherlands
            gdf_filtered.plot(ax=ax, color='blue', alpha=0.5)

            # Show the plot
            plt.title('GeoJSON Data within the Netherlands')
            plt.xlabel('Longitude')
            plt.ylabel('Latitude')
            plt.show()
        else:
            print("No 'features' found in the GeoJSON data.")

    else:
        print(f"Failed to get data from API. Response Code: {response.getcode()}")

except Exception as e:
    print(f"Error occurred: {e}")
```
